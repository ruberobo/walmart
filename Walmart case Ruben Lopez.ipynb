# _Ejercicio 1_

```python
import pandas as pd
import numpy as np
```

```python
asignaturas=pd.read_csv('Asignaturas.csv')
estudiantes=pd.read_csv('estudiantes.csv')
inscripcion=pd.read_csv('Inscripcion.csv')
```

```python
asignaturas
```

```python
pd1=inscripcion.merge(asignaturas, on='asignatura_id')
```

```python
pd1
```

```python
pd2=pd1.merge(estudiantes,how='left', left_on='num_cuenta', right_on='numero_cuenta', suffixes=('_materia','_alumno'))
```

```python
pd2
```

```python
promedios=pd2.pivot_table(index='asignatura_id', values='calificacion', aggfunc='mean').reset_index()
```

```python
df=pd2.merge(promedios, how='left', on='asignatura_id', suffixes=('','_promedio'))
```

```python
df['calificacion']=df['calificacion'].astype(float)
```

```python
df['bandera']=np.where(df['calificacion']>df['calificacion_promedio'],1,0)
```

```python
df.columns
```

```python
df_exp=df[['nombre_alumno','nombre_materia','calificacion','calificacion_promedio', 'bandera']]
```

```python
df_exp.to_csv('ejercicio_1.csv', sep=',', index=False)
```
